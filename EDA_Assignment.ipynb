```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

weather = pd.read_csv('/content/weather_forecast_data.csv')
weather.head()
```

```python
#give the stripplot between Rain and temperature
sns.stripplot(x='Rain', y='Temperature', data=weather)

plt.show()
```

```python
sns.boxplot(x='Rain',y='Temperature',data=weather) #boxplot

plt.show()
```

```python
d = np.random.randint(1,100,(5,5))

sns.heatmap(d,annot=True,cbar=True,color='green')

plt.show()
```

```python
sns.barplot(x='Rain',y='Temperature',data=weather,color="green")

plt.xlabel('Rain')
plt.ylabel('Temperature')
plt.title('Rain vs Temperature')

plt.show()
```

```python
sns.histplot(weather['Temperature'],kde=True,color='orange')

plt.show()
```

```python
# disturbution of temperature,humidity,wind sped, cloud cover,pressure
weather.hist(edgecolor='black',bins = 25, figsize= (12,10),color="red")
plt.show()
```

```python
sns.pairplot(data = weather, hue="Rain", size=3)
plt.show()
```
