4)  Histogram

o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.

o Customize the number of bins and add a grid to the plot.

```python
data = np.random.randn(100)
plt.hist(data, bins=20, color='orange', edgecolor='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Values')
plt.grid(True)
plt.show()

```

8)  Comparison of Matplotlib and Seaborn

o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.

o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]

plt.bar(students, marks, color='orange')
plt.title("Matplotlib BarPlot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

print("---------------------------------------------------------------")

sns.barplot(x=students, y=marks)
plt.title("Seaborn BarPlot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

9) Pair Plot (Seaborn)

o Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import pandas as pd

data = {
    'Age': [22, 25, 30, 35, 40, 45, 50, 55, 60, 65],
    'Height': [165, 170, 175, 180, 185, 190, 195, 200, 205, 210],
    'Weight': [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
}
df = pd.DataFrame(data)
sns.pairplot(df)
plt.title('Pair Plot of Age, Height, and Weight')
plt.show()
```
